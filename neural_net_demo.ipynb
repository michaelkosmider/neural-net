imports

```python
import neural_net
import tensorflow as tf #tensorflow is only being imported for easy access to the fashion mnist dataset
```

obtaining and processing the fashion mnist dataset

```python
fashion_mnist = tf.keras.datasets.fashion_mnist.load_data()
(X_train_full, y_train_full), (X_test, y_test) = fashion_mnist
X_train, y_train = X_train_full[:-5000], y_train_full[:-5000]
X_valid, y_valid = X_train_full[-5000:], y_train_full[-5000:]

X_train = X_train.reshape((55000, 784)) 
X_valid = X_valid.reshape((5000, 784))
X_train = X_train/255
X_valid = X_valid/255
```

training the neural network

```python
model = neural_net.Neural_net(shape=[300,100], epochs=300, learning_rate=0.01)
model.fit(X_train, y_train, X_valid, y_valid)
```
